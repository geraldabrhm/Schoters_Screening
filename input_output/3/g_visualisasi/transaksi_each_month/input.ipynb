```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
df = pd.read_csv('../../c_total_transaction/each_month/output.csv')
df['Bulan_Tahun'] = pd.to_datetime(df['Bulan_Tahun'], format='%B %Y')
df['Bulan_Tahun'] = df['Bulan_Tahun'].dt.strftime('%B-%Y')
```

```python
df_to_plot = pd.DataFrame()
df_to_plot['Transaksi'] = df['Total_Transaksi']
df_to_plot = df_to_plot.set_index(df['Bulan_Tahun'])
plt.ticklabel_format(style='plain')
plt.plot(df_to_plot)
plt.gcf().autofmt_xdate()
plt.savefig("output.png")
```
